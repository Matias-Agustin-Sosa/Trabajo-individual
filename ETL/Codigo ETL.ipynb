```python
import pandas as pd
import numpy as np
import ast
```

```python
# trabajo el dataset movies
df_movies = pd.read_csv(r"C:\Users\matia\OneDrive\Escritorio\Henry\Trabajo individual\movies_dataset.csv")
```

```python
df_movies.shape
```

```python
# vemos valores nulos
df_movies.isnull().sum()
```

```python
# remplazamos por 0 los nulos en las columnas deseadas
df_movies[["revenue", "budget"]] = df_movies[["revenue", "budget"]].fillna(0)
df_movies[["revenue", "budget"]].isnull().sum()
```

```python
# Borro filas con nulos
df_movies.dropna(subset = ["release_date"], inplace=True)
df_movies["release_date"].isnull().sum()
```

```python
# filas con error en las fechas
df_movies["release_date"].loc[[19730,29503,35587]]
```

```python
# las filas con error en la fecha quedan como nulos
df_movies["release_date"] = pd.to_datetime(df_movies["release_date"], format = "%Y-%m-%d", errors = "coerce")
df_movies["release_date"]
```

```python
# valores_nulos = df_movies["release_date"].isnull()
# df_filtrado = df_movies[valores_nulos]
# print(df_filtrado)
```

```python
# Borro filas con nulos
df_movies.dropna(subset = ["release_date"], inplace=True)
df_movies["release_date"].isnull().sum()
```

```python
# miro el tipo de dato de cada columna
print(df_movies.dtypes)
```

```python
# creo una columna extrayendo el a√±o de release_date
df_movies["release_year"] = df_movies["release_date"].dt.year
df_movies.info()
```

```python
# combierto la columna budget a flotante
df_movies["budget"] = df_movies["budget"].astype(float)
```

```python
# veo nulos
df_movies["budget"].isnull().sum()
```

```python
# calculo la ganancia por pelicula, en caso de no conocer el costo de produccion lo remplazo por 0 
df_movies["return"] = np.where(df_movies["budget"] == 0, 0, df_movies["revenue"] / df_movies["budget"])
df_movies["return"].info()
```

```python
# elimino columnas inecesarias
eliminar = ["video","imdb_id","adult","original_title","poster_path", "homepage"]
df_movies.drop(eliminar, axis = 1, inplace = True)
```

```python
# reviso columnas y su tipo de dato
df_movies.info()
```

```python
# tama√±o original
df_movies.shape
```

```python
# tama√±o sin duplicados 
df_movies = df_movies.drop_duplicates()
df_movies.shape
```

```python
df_movies
```

```python
# desanidamiento de columnas "[{}]"
df_movies["genres"] = df_movies["genres"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["name"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["production_companies"] = df_movies["production_companies"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["name"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["production_countries_name"] = df_movies["production_countries"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["name"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["production_countries_iso"] = df_movies["production_countries"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["iso_3166_1"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["spoken_languages_name"] = df_movies["spoken_languages"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["name"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["spoken_languages_iso"] = df_movies["spoken_languages"].fillna("[]").apply(ast.literal_eval).apply(lambda x: [i["iso_639_1"] for i in x] if isinstance(x,list) else [])
```

```python
df_movies["belongs_to_collection"] = df_movies["belongs_to_collection"].fillna("{}").apply(ast.literal_eval)
```

```python
def extraer(x):
    return x["name"] if "name" in x else "Sin Dato"

df_movies["belongs_to_collection"] = df_movies["belongs_to_collection"].apply(extraer)
```

```python
# Elimino columnas con valores repatidos
columnas_a_eliminar = ["spoken_languages", "production_countries"]
df_movies.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
df_movies
```

```python
# creamos el data set nuevo
df_movies.to_csv('Movies_dataset(ETL).csv')
```
